```python
# remove columns empty columns
df = df.drop(columns=['Column 1', 'Column 2', 'Unnamed: 7', 'Unnamed: 9'])
```

```python
#Converting the "Order Date" and "Ship Date" columns to datetime format
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
#creating new column to ectract month from order date
df["Order Month"] = df["Order Date"].dt.month
```

```python
#Grouping the data by "Order Month" and "Category" to get the total sales for each category by month
monthly_sales = df.groupby(["Order Month", "Category"])["Sales"].sum().reset_index()
```

```python
#Pivoting the monthly sales data to create a new dataframe where the rows are the months and the columns are the categories
monthly_sales_pivot = monthly_sales.pivot(index="Order Month", columns="Category", values="Sales")
```

```python
print(monthly_sales_pivot)
```

```python
#the percentage of total sales for each category by month
monthly_sales_pivot["Total"] = monthly_sales_pivot.sum(axis=1)
monthly_sales_percentage = monthly_sales_pivot.apply(lambda x: 100*x/x["Total"], axis=1).drop("Total", axis=1)
```

```python
#Creating a new column "Days to Ship" to calculate the number of days between the "Order Date" and "Ship Date"
df["Days to Ship"] = (df["Ship Date"] - df["Order Date"]).dt.days
```

```python
df_monthly = df.resample('M').sum()
```

```python
df_monthly.plot(figsize=(12,6))
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.show()
```

```python
# Testing the stationarity of the time series
result = sm.tsa.stattools.adfuller(df_monthly['Sales'])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

```python
# Choosing the ARIMA model parameters
fig, ax = plt.subplots(2, 1, figsize=(10,8))
sm.graphics.tsa.plot_acf(df_monthly['Sales'], lags=12, ax=ax[0])
sm.graphics.tsa.plot_pacf(df_monthly['Sales'], lags=12, ax=ax[1])
plt.show()
```

```python
model = sm.tsa.ARIMA(df_monthly['Sales'], order=(1,1,1))
results = model.fit()
```

```python
print(results.summary())
residuals = results.resid
print(residuals.describe())
```

```python
pred_start_date = df_monthly.index[-1] + pd.DateOffset(months=1)
pred_end_date = pred_start_date + pd.DateOffset(years=1)
pred = results.predict(start=pred_start_date, end=pred_end_date, dynamic=True)
```

```python
# Visualize the predictions
df_monthly['Sales'].plot(figsize=(12,6))
pred.plot()
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.title('Monthly Sales Predictions')
plt.show()
```
